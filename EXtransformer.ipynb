```python
from transformers import AutoTokenizer

english_tokenizer = AutoTokenizer.from_pretrained("roberta-base")
thai_tokenizer = AutoTokenizer.from_pretrained("airesearch/wangchanberta-base-att-spm-uncased")
```

```python
src_dict=english_tokenizer(en, padding = 'max_length', max_length = 256)
src,src_mask=torch.tensor(src_dict['input_ids']).to(torch.int64),torch.tensor(src_dict['attention_mask']).bool()

tgt_dict=thai_tokenizer(th, padding = 'max_length', max_length = 256)
tgt,tgt_mask=torch.tensor(tgt_dict['input_ids']).to(torch.int64),torch.tensor(tgt_dict['attention_mask']).bool()
```

```python
from torch.utils.data import DataLoader, Dataset

class Dataset(Dataset):
    def __init__(self, src, src_mask, tgt, tgt_mask):
        super().__init__()
        self.src = src
        self.mask_src = src_mask
        self.tgt = tgt
        self.mask_tgt = tgt_mask

    def __getitem__(self, index):
        return self.src[index,:], self.mask_src[index,:], self.tgt[index,:], self.mask_tgt[index,:]

    def __len__(self):
        return self.src.shape[0]
```

```python
import pytorch_lightning as pl
from datasets import load_metric
import deepcut

metric = load_metric("sacrebleu")
```

```python
def deepcut_add_space(target_list):
  result=[]
  for sequence in target_list:
    sequence_token=deepcut.tokenize(sequence)
    result.append([' '.join(sequence_token)])
  return result
```

```python
class LitModel(pl.LightningModule):
  def __init__(self):
    super().__init__()
    self.model=XTransformer(
    dim = 512,
    #encode รับ eng
    enc_num_tokens = len(english_tokenizer.vocab)+100,
    enc_depth = 6,
    enc_heads = 8,
    enc_max_seq_len = 1024,
    #decoder รับ thai
    dec_num_tokens = len(thai_tokenizer.vocab)+100,
    dec_depth = 6,
    dec_heads = 8,
    dec_max_seq_len = 1024,
    tie_token_emb = True      # tie embeddings of encoder and decoder
)
  def forward(self, x):
    #คาดหวังว่า x จะเป็น list ของ sequence ที่เป็นภาษาอังกฤษ
    src_dict=english_tokenizer(x, padding = 'max_length', max_length = 1024)
    src,src_mask=torch.tensor(src_dict['input_ids']).to(torch.int64),torch.tensor(src_dict['attention_mask']).bool()

    start_tokens = (torch.ones((src.shape[0], 1)) * 1).long()

    genrated_data = self.model.generate(src , start_tokens, 1024, src_mask=src_mask)
    result = thai_tokenizer.batch_decode(genrated_data, skip_special_tokens = True)

    result = deepcut_add_space(result)

    return result

  def training_step(self, batch, batch_idx):
    src, mask_src, tgt, mask_tgt=batch
    loss = self.model(src, tgt, src_mask = mask_src, tgt_mask = mask_tgt)
    self.log("train_loss", loss)
    return loss
    
  def validation_step(self, batch, batch_idx):
    src, mask_src, tgt, mask_tgt=batch
    start_tokens = (torch.ones((src.shape[0], 1)) * 1).long()
    genrated_data = self.model.generate(src, start_tokens, 256, src_mask=mask_src)
    
    prediction = thai_tokenizer.batch_decode(genrated_data, skip_special_tokens = True)
    reference = thai_tokenizer.batch_decode(tgt, skip_special_tokens = True)

    prediction = deepcut_add_space(prediction)
    reference = deepcut_add_space(reference)

    BLEU = metric.compute(prediction, reference)

    self.log('valid_BLEU',BLEU)
    return BLEU
  
  def configure_optimizers(self):
    optimizer = torch.optim.AdamW(self.model.parameters(),lr=0.001)
    return optimizer
```

```python
train, val = Dataset(src[:9000], src_mask[:9000], tgt[:9000], tgt_mask[:9000]), Dataset(src[9000:10000], src_mask[9000:10000], tgt[9000:10000], tgt_mask[9000:10000])

model=LitModel()
trainer = pl.Trainer()
trainer.fit(model, DataLoader(train, batch_size=16, num_workers=1), DataLoader(val, batch_size=16, num_workers=1))
```
